# Spam Mail Prediction

```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Pre-Processing

```python
# Loading the data from csv file to a pandas DataFrame
df=pd.read_csv('/content/mail_data.csv')
df.head()
```

```python
df.shape
```

In this dataset included 5572 number of raws and 2 columns.

```python
df.info()
```

By this information there is no any null values.

Label Encoding

```python
# Label spam mail as 0, ham mail as 1
df.loc[df['Category']=='spam','Category',] = 0
df.loc[df['Category']=='ham','Category',] = 1
```

spam - 0  ham - 1

```python
from re import X
# Separating the data as texts and label
X = df['Message']
Y = df['Category']  # This is response variable that is what we predict

```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature extraction

```python
# Transform the text data to feature vectors that can be used as input to the Logistic regression
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

min_df - Choose all the words that have  more than one wants,stop_words it will be remove that we won't that words.

```python
print(X_train)
```

```python
print(X_train_features)
```

Training the Model

Logistic Redression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

Evaluating the trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features) # predicted values
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

It is compare the real response value and the predicted response value and checking the accuracy.

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

96% of accuracy which is very good that means our model working as really well.

```python
# prediction on test data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

Building a Predictive System

```python
input_mail = ["Had your mobile 11 months or more? U R entitled to Update to the latest colour mobiles with camera for Free! Call The Mobile Update Co FREE on 08002986030"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```

The results demonstrate that the proposed spam mail prediction system using TF-IDF vectorization and Logistic Regression is accurate, reliable, and efficient. With a testing accuracy of 96.68%, the model successfully classifies emails as spam or non-spam with high reliability. Therefore, this approach is suitable for real-world spam detection applications, though further improvements can be made by incorporating advanced models and additional evaluation metrics.
